```python
import numpy as np
import pandas as pd
import matplotlib as plt
import seaborn as sns
%matplotlib inline
```

```python
df_matches=pd.read_csv('matches.csv')
df_matches.head()
```

```python
df_player=pd.read_csv('player.csv')
df_player.head()
```

```python
df_player_attributes=pd.read_csv('player_attributes.csv')
df_player_attributes.head()
```

```python
df_team_attributes=pd.read_csv('team_attributes.csv')
df_team_attributes.head()
```

```python
df_teams=pd.read_csv('teams.csv')
df_teams.head()
```

```python
df_countries=pd.read_csv('countries.csv')
df_countries.head()
```

```python
df_leagues=pd.read_csv('leagues.csv')
df_leagues.head()
```

```python
df_countries_modified=df_countries[df_countries['name']=='England'].append([df_countries[df_countries['name']=='France'],df_countries[df_countries['name']=='Germany'],df_countries[df_countries['name']=='Italy'],df_countries[df_countries['name']=='Spain']])
df_countries_modified.columns=['index','country_id','country']
df_countries_modified
```

```python
df_countries_leagues=pd.merge(df_countries_modified,df_leagues,how='inner',on='country_id')
df_countries_leagues=df_countries_leagues.drop('id.1',axis=1).drop('name.1',axis=1).drop('id',axis=1).drop('Unnamed: 0',axis=1).drop('index',axis=1)
df_countries_leagues.columns=['country_id','country','league_name']
df_countries_leagues
```

```python
df_matches_modified=df_matches.__getitem__(['season','country_id','home_team_goal','away_team_goal'])
df_matches_modified
```

```python
df_countries_leagues_seasons=pd.merge(df_countries_leagues,df_matches_modified,how='inner',on='country_id')
df_countries_leagues_seasons
```

```python
df_teams_modified1=df_teams.drop('Unnamed: 0',axis=1).drop('id',axis=1).drop('team_fifa_api_id',axis=1).drop('team_short_name',axis=1)
df_teams_modified1.columns=['home_team_api_id','team_name']
df_teams_modified1
```

```python
df_teams_modified2=df_teams.drop('Unnamed: 0',axis=1).drop('id',axis=1).drop('team_fifa_api_id',axis=1).drop('team_short_name',axis=1)
df_teams_modified2.columns=['away_team_api_id','team_name']
df_teams_modified2
```

```python
df_teams_modified3=pd.merge(df_teams_modified1,df_matches,how='inner',on='home_team_api_id')
df_teams_modified3
```

```python
df_teams_modified4=pd.merge(df_teams_modified2,df_matches,how='inner',on='away_team_api_id')
df_teams_modified4
```

```python
df_team_name=df_teams_modified3+df_teams_modified4
df_team_name_modified=df_team_name.__getitem__(['country_id','team_name'])
df_team_name_modified
```

Final Dataset (Too big for RAM to process):

```python
df_final=pd.merge(df_countries_leagues_seasons,df_team_name_modified,how='inner',on='country_id')
df_final
```
